```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
A = np.array([[74.00, 80.00, 84.00, 90.50], [71.00, 74.50, 76.00, 77], [81.00, 85.00, 88.00, 86.00], [90.00, 95.00, 97.00, 98.00], ['35%', '25%', '15%', '25%']])
A[0:4] = A[0:4].T
A
```

```python
Adf = pd.DataFrame(A, index = ['June', 'July', 'August', 'September', 'Weight'], columns = ['External Vuln Health Score', 'Web Vulnerability Health Score', 'Patch Management Health Score', 'Anti-virus Health Score'])
```

```python
Adf
```

```python
Adf.iloc[0:4] = Adf.iloc[0:4].apply(pd.to_numeric)
```

```python
Adf['Total Health Index'] = Adf['External Vuln Health Score'][0:4]*0.35 + Adf['Web Vulnerability Health Score'][0:4]*0.25 + Adf['Patch Management Health Score'][0:4]*0.15 + Adf['Anti-virus Health Score'][0:4]*0.25
```

```python
Adf
```

```python
plt.figure(figsize =(10, 10))
plt.plot(Adf['External Vuln Health Score'][0:4], 'p--', label='External Vuln Health Score')
plt.plot(Adf['Web Vulnerability Health Score'][0:4], 'g--', label='Web Vulnerability Health Score')
plt.plot(Adf['Patch Management Health Score'][0:4], 'b--', label='Patch Management Health Score')
plt.plot(Adf['Anti-virus Health Score'][0:4], 'y--', label='Anti-virus Health Score')
plt.plot(Adf['Total Health Index'][0:4], 'r', label='Total Health Index')

plt.title('Information Scurity Health Index')
plt.xlabel('Score')
plt.ylabel('Month')
plt.legend()
plt.show()
```

```python
values = [99.0, 99.3, 93.0, 99.4, 95.0, 96.465]
values1 = [74, 99.2, 90, 98,65,81.98]
```

```python
features = ['Password Strength Score', 'Awareness Index', 'Laptop Encryption Compliance', 'Software Compliance', 'User Behavior Testing','Total Behavior Index']
```

```python
angles=np.linspace(0, 2*np.pi, len(values), endpoint=False)
angles
```

```python
values=np.concatenate((values,[values[0]]))
angles=np.concatenate((angles,[angles[0]]))
values1=np.concatenate((values1,[values1[0]]))
fig=plt.figure()
# 设置为极坐标格式
ax = fig.add_subplot(111, polar=True)
# 绘制折线图
ax.plot(angles, values, 'b-', linewidth=2, label='June')
ax.plot(angles, values1, 'r-', linewidth=2, label='September')
ax.fill(angles, values1, 'r', alpha=0.15)
# 填充颜色
#ax.fill(angles, values, alpha=0.15)
 
# # 设置图标上的角度划分刻度，为每个数据点处添加标签
ax.set_thetagrids(angles * 180/np.pi, features)
 
# # 设置雷达图的范围
#ax.set_ylim(0,6)
# # 添加标题
plt.title('Behavior Index')
# # 添加网格线
plt.legend()
plt.show()
 
```

```python
C = pd.DataFrame([['30%', '40%', '30%']], index=['weight'], columns= ['Vulnerability Management Processes', 'User Provisioning Processes', 'Disaster Recovery Processes'])
```

```python
Adf = Adf.append(C)
Adf
```

```python
Adf['Total Process Index'] = Adf['Vulnerability Management Processes'][0:4] * 0.3 + Adf['User Provisioning Processes'][0:4] * 0.4+ Adf['Disaster Recovery Processes'][0:4] * 0.3 
```

```python
Adf
```

```python
label_list = ['Vulnerability Management Processes', 'User Provisioning Processes', 'Disaster Recovery Processes']
a = Adf['Vulnerability Management Processes'][0:4].mean()
b = Adf['User Provisioning Processes'][0:4].mean()
c = Adf['Disaster Recovery Processes'][0:4].mean()
size = [a,b,c]
color = ["red", "green", "blue"]
plt.figure(figsize=(10, 10))
plt.pie(size, colors=color, labels=label_list, labeldistance=1.1, autopct="%1.1f%%", shadow=False, startangle=90, pctdistance=0.6)
plt.axis("equal")    # 设置横轴和纵轴大小相等，这样饼才是圆的
plt.legend()
plt.show()
```
